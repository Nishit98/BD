```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nodes_df = pd.read_csv('nodecsv.csv')
edges_df = pd.read_csv('edges.csv')

G = nx.DiGraph()

#add nodes to the graph
for index , row in nodes_df.iterrows():
    G.add_node(row['Node_ID'],name=row['Name'])

#addedged to the graph
for index, row in edges_df.iterrows():
    G.add_edge(row['Source_Node'],row['Target_Node'],
              matches_together = row['Matches_Together'],
              partnership_runs = row['Partnership_Runs'],
              wickets_shared=row['Wickets_Shared'])

degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
pagerank = nx.pagerank(G)

print("degree_centrality :" ,degree_centrality)
print("betweenness_centrality :",betweenness_centrality)
print("clossness_centrality :" , closeness_centrality)
print("pagerank :",pagerank)
```

```python
# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)  # Layout for better visualization
nx.draw(G, pos, with_labels=True, node_size=1800, font_size=10, font_color='black', font_weight='bold')

# Draw edge labels for key attributes (you can customize which attribute to display)
edge_labels = nx.get_edge_attributes(G, 'matches_together')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title('Social Network Graph')
plt.show()
```
